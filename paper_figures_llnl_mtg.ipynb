```python
from scipy.stats import lognorm

def gauss(x, A, mu, sig):
    return A*np.exp(-(x-mu)**2/(2*sig**2))

def multigauss(x, A1, m1, s1, A2, m2, s2, sh):
    return gauss(x, A1, m1, s1) + A2*lognorm.pdf((x-m2)/s2, sh)
```
